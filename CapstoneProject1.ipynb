# 4. Dropping irrelevant columns

```python
df_data=df_data.drop(['Date', 'DrugId', 'Reviews', 'Sides'], axis=1)
```

# 5. Renaming the columns

```python
df_data=df_data.rename(columns={"Satisfaction": "Rating", "UsefulCount":"Count"})
```

# 6. Dropping the Irrelevant rows

```python
#selecting the rows of our target which is a condition only have pain 
df_data=df_data[df_data.Condition.str.contains('Pain', na=False)]
```
